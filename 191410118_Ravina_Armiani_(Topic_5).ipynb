**Anomalies and Outlier Handling**

```python
#mencari outlier pada data tenure
#mencari nilai kuartil 1 dan kuartil 3
Q1 = df['tenure'].quantile(0.25)
Q3 = df['tenure'].quantile(0.75)
Q1,Q3
```

```python
#mencari IQR (selisih anatara kuartil 3 dan kuarti 1) pada data tenure
IQR = Q3 - Q1
IQR
```

```python
#mencari nilai batas terendah dan nilai batas tertinggi pada data tenure
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
lower_bound, upper_bound
```

```python
#mencari outliers
outliers = df[df['tenure']>upper_bound]
outliers
```
